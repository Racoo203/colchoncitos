```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('C:/Users/riosz/OneDrive/Documentos/GitHub/colchoncitos/WorkingDB2.csv')
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%y')
```

```python
dfFinal = pd.DataFrame({
    'Producto': [],
    'Lugar': [],
    'Precio Unitario': [],
    'Cantidad': [],
    'Fecha': []
})
```

```python
def AgregarRegion(df_fin,df_encoded,Region):
    ventas = Region + ' total $'
    cantidad = Region + ' ctdad'
    Producto = df_encoded['Encoded Products']
    Cantidad = df_encoded[cantidad] 
    Fecha = df_encoded['date']
    Precio_Unitario = df_encoded[ventas]
    provicional = pd.DataFrame({
        'Producto': Producto,
        'Lugar': [Region]*len(Cantidad),
        'Precio Unitario': Precio_Unitario,
        'Cantidad': Cantidad,
        'Fecha': Fecha
    })
    final = pd.concat([df_fin, provicional], axis = 0)
    return final
```

```python
regiones = ['Cadereyta', 'Coahuila', 'Durango', 'Expos F', 'Laredo', 'Monterrey', 'Online', 'Queretaro', 'Reynosa', 'San Luis']
for i in regiones:
    dfFinal = AgregarRegion(dfFinal, df, i)
```

```python
dfFinal['Cantidad'] = dfFinal['Cantidad'].replace(0,1)
dfFinal['Precio Unitario'] = dfFinal['Precio Unitario']/dfFinal['Cantidad']
dfFinal['Cantidad'] = dfFinal['Cantidad'].replace(1,0)
```
