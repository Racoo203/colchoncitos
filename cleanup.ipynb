```python
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
data = pd.read_csv('encoded_db-1.csv')
data.head()
```

```python
data.columns
```

```python
byebye = []
for column in data:
    if data[column].dtype == 'object':
        data[column] = data[column].str.strip()
    if ('Unnamed' in column):
        byebye.append(column)

data = data.drop(byebye, axis=1)
data.head()
```

```python
data.drop(['index', 'Total libre de impuestos', 'Ctdad Ordenada', 'Indefinido ctdad'], axis=1, inplace=True)
data.head()
```

```python
data = data.melt(['date', 'Encoded Products'])

data['variable'] = data['variable'].str.replace(' $','')

data['Store'] = data['variable'].str[:-5]
data['variable'] = data['variable'].str[-5:]

month_encoding = (
    'ene', 'feb', 'mar', 'abr', 'may', 'jun', 'jul', 'ago', 'sep', 'oct', 'nov', 'dic'
)

data['Day'] = data['date'].str.split(' ').str[0].astype(int)
data['Month'] = data['date'].str.split(' ').str[1].apply(lambda x: month_encoding.index(x) + 1)
data['Year'] = data['date'].str.split(' ').str[2].astype(int)

data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']])
data.drop('date', axis=1, inplace=True)

data['Day of Week'] = data['Date'].dt.dayofweek
data = data.loc[data['variable'] == 'ctdad']
data.drop(['variable'], axis=1, inplace=True)
data.rename(columns={'value': 'Units'}, inplace=True)

data.reset_index(drop= True, inplace=True)
```

```python
for column in data:
    if data[column].dtype == 'object':
        data[column] = data[column].str.strip()
```

```python
print(data.isna().any())
data.dropna(inplace=True)
```
