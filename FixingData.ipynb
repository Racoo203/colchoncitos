```python
import pandas as pd
import numpy as np
import copy
import locale

from datetime import datetime as dt
locale.setlocale(locale.LC_TIME,'es_ES.UTF-8')
df_orig = pd.read_excel("ventas.xls", header=0)
df_working = copy.deepcopy(df_orig)
```

```python
df_orig.iloc[:50,:]
```

```python
dates = []
indeces = []
for i, obj in enumerate(df_orig["Unnamed: 0"]):
    if type(obj) is str:
        if len(obj) > 15:
            dates.append(obj)
            indeces.append(i)
            df_working["Unnamed: 0"][i] = np.nan
            
            
print(dates)
print(indeces)
```

```python
dates_col = np.zeros_like(df_orig["Unnamed: 0"])
dates_col[:indeces[0]] = np.nan

for i in range(1, len(indeces)):
    dates_col[indeces[i-1]:indeces[i]] = dates[i-1][5:].replace(" ", ".")
    
dates_col[indeces[-1]:] = dates[-1][5:].replace(" ", ".")

df_working["Unnamed: 0"][0] = "Region"
df_working["Unnamed: 0"][1] = "Titulo"
    


        
```

```python
df_working["Dates"] = dates_col
df_working[["Dates","Unnamed: 0"]].iloc[70:,:]
```

```python
df_working.iloc[0:,1:].fillna(0, inplace=True)
df_working.dropna(inplace=True,axis=0)
for i in range(len(df_working.columns)-3):
    if df_working.iloc[0,i] == 0:
        df_working.iloc[0,i] = df_working.iloc[0,i-1]
        #df_working.iloc[0,i-1] = np.nan
```

```python
#df_working.dropna(axis=1, inplace=True)
df_working.drop(["Unnamed: 21","Unnamed: 22"], axis=1, inplace=True)
```

```python
df_working.columns = df_working.iloc[0,:].astype(str) + " " + df_working.iloc[1,:].astype(str)
df_working.rename(columns={"Region Titulo":"Producto","0 0" : "Fecha"}, inplace=True)
```

```python
df_working.drop([0,1,2], inplace=True)
```

```python
df_working.reset_index(drop=True, inplace=True)
df_working['Fecha'] = pd.to_datetime(df_working['Fecha'], format='%d.%b%Y')
```

```python
def div(numerator, denominator):
  return lambda row: 0.0 if row[denominator] == 0 else float(row[numerator]/row[denominator])
```

```python
df_working.columns = df_working.columns.str.replace("Total libre de impuestos", "Precio por unidad")
df_working.columns = df_working.columns.str.replace("Ctdad Ordenada", "Cantidad ordenada")
```

```python
for i in range(1, len(df_working.columns)-1,2):
    df_working[df_working.columns[i]] = df_working.apply(div(df_working.columns[i], df_working.columns[i+1]), axis=1)
    
```
