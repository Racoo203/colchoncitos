```python
import pandas as pd
import numpy as np
import copy
df_orig = pd.read_excel("ventas.xls")
df_working = copy.deepcopy(df_orig)
```

```python
df_orig.iloc[:50,:]
```

```python
dates = []
indeces = []
for i, obj in enumerate(df_orig["Unnamed: 0"]):
    if type(obj) is str:
        if len(obj) > 15:
            dates.append(obj)
            indeces.append(i)
            
print(dates)
print(indeces)
```

```python
dates_col = np.zeros_like(df_orig["Unnamed: 0"])
dates_col[:indeces[0]] = np.nan

for i in range(1, len(indeces)):
    dates_col[indeces[i-1]:indeces[i]] = dates[i-1]
    
dates_col[indeces[-1]:] = dates[-1]

    
    


        
```

```python
df_working["Dates"] = dates_col
```
