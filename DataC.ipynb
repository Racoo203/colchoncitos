### DF regiones

```python
dfCa = df[['Encoded Products','Cadereyta ctdad','Cadereyta total $','date']]
dfCa['Region'] = 'Cadereyta'

dfCo = df[['Encoded Products','Coahuila ctdad','Coahuila total $','date']]
dfCo['Region'] = 'Coahuila'

dfDu = df[['Encoded Products','Durango ctdad','Durango total $','date']]
dfDu['Region'] = 'Durango'

dfEF = df[['Encoded Products','Expos F ctdad','Expos F total $','date']]
dfEF['Region'] = 'Expos F'

dfLa = df[['Encoded Products','Laredo ctdad','Laredo total $','date']]
dfLa['Region'] = 'Laredo'

dfMty = df[['Encoded Products','Monterrey ctdad','Monterrey total $','date']]
dfMty['Region'] = 'Monterrey'

dfOn = df[['Encoded Products','Online ctdad','Online total $','date']]
dfOn['Region'] = 'Online'

dfQro = df[['Encoded Products','Queretaro ctdad','Queretaro total $','date']]
dfQro['Region'] = 'Queretaro'

dfRy = df[['Encoded Products','Reynosa ctdad','Reynosa total $','date']]
dfRy['Region'] = 'Reynosa'

dfSL = df[['Encoded Products','San Luis ctdad','San Luis total $','date']]
dfSL['Region'] = 'San Luis'

dfSL.head()
```

```python
DF = [dfCa, dfCo, dfDu, dfEF, dfLa, dfMty, dfOn, dfQro, dfRy, dfSL]

for df in DF:
    df.columns = ['Producto', 'Cantidad', 'Total', 'Fecha','Region']
```

```python
dfF = pd.concat(DF, axis=0)
```

```python
dfF['Cantidad'] = dfF['Cantidad'].replace(0,1)
dfF['Precio Unitario'] = dfF['Total']/dfF['Cantidad']
dfF['Cantidad'] = dfF['Cantidad'].replace(1,0)
```

```python
dfF = dfF.reindex(columns=['Producto', 'Region', 'Precio Unitario', 'Cantidad', 'Fecha'])

dfF.sample(20)
```
